```python
# Import testing CSV

import pandas as pd

ala_results = pd.read_csv("results_no_head.csv", header=None, names=['label', 'title'] ,sep=",")

ala_results.head(10)
```

```python
guy_results = pd.read_csv("predictions.csv", header=None, names=['label', 'title'] ,sep=",")

guy_results.head(10)
```

```python
amb_results = pd.read_csv("new_data_amb.csv", header=None, names=['label', 'title', 'process'] ,sep=",")

amb_results = amb_results.drop(columns=['process'])

amb_results.head(10)
```

```python
# Count the occurrences of each unique value in the 'label' column
label_counts_ala = ala_results['label'].value_counts()#/9984*100
label_counts_guy = guy_results['label'].value_counts()#/9984*100
label_counts_amb = guy_results['label'].value_counts()#/9984*100

# Print the counts
print(label_counts_ala)
print(label_counts_guy)
print(label_counts_amb)
```

```python
# Combine the datasets into a single DataFrame for comparison
comparison = pd.DataFrame({
    'title': ala_results['title'],           # Titles from ala_results (assumed identical across datasets)
    'ala_label': ala_results['label'],       # Labels from ala_results
    'guy_label': guy_results['label'],       # Labels from guy_results
    'amb_label': amb_results['label']      # Labels from the third dataset
})

# Find rows where at least one label disagrees
disagreement = comparison[
    (comparison['ala_label'] != comparison['guy_label']) |
    (comparison['ala_label'] != comparison['amb_label']) |
    (comparison['guy_label'] != comparison['amb_label'])
]

# Take a random sample of rows where predictions disagree
sample_disagreement = disagreement.sample(n=20, random_state=4)  # Adjust n for the sample size

# Display the sample as a DataFrame with full column width
pd.set_option('display.max_colwidth', None)  # Allow full display of the 'title' column
sample_disagreement.head(30)
```
